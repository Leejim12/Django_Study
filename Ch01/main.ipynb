### HTTP (Hypertext Transfer Protocol)
웹 서버 - 클라이언트 사이에서 데이터 주고받는 규약
- Based on TCP/IP Protocol.
- Hypertext 외에도, 컴퓨터에서 다룰 수 있는 모든 데이터 전송 가능.

### HTTP Method


get == select
post == create
put == update
Delete == delete

head == 메타데이터 취득
Options == 메소드 취득
Trace == 루프백 시험
connect == 프록시 동작 터널 접속으로 변경

### GET , POST
HTML 폼에서 지정할 수 있는 메소드는 위 두개 뿐.
- get
?이름=값
```
http://docs.djangoproject.com/search/?q=forms~~
```

- POST
요청 메시지의 Body에 데이터 넣어 보냄
```
POST https://docs.djangoproject.com/ssearch/HTTP/1.1
Content-Type: application/x-www-form-urlencoded

q = forms&release = 1
```
* url 은 길이제한이 있음
=> 보안의 이유와 추가로, post가 유리함.

### HTTP상태코드
1. 1xx
Informational : 정보제공
```
임시 응답.
현재 클라이언트 요청까지는 처리 완료. 계속 진행
(HTTP 1.1 추가)
```

2. 2xx
Success : 성공
```
클라이언트 요청 서버에서 처리 성공
```

3. 3xx
Redirection : 리다이렉션
```
완벽한 처리를 위해 추가 동작 필요
대부분 서버 주소, url 문서 이동하였으니 그 주소로 다시 시도하라는 의미
```

4. 4xx
Client Error : 클라이언트 에러
```
없는 페이지 요청 등.
요청 측의 에러
```

5. 5xx
Server Error : 서버 에러
```
서버 사정으로 메시지 처리에 문제 발생.
=> 서버 부하, DB처리 오류, 서버 Exception 발생 등
```

### 정적 vs 동적 페이지
- 정적
: 요청 시점, 사람이 달라도 항상 같은 내용 표시해줌.

- 동적
: 동일한 resource 요청이라도, 누가 언제 어떻게 요구했는지에 따라 다른 내용이 표시되는 페이지.
ex. 현재 시각
    쇼핑몰 장바구니
* CGI : 프로그램과 웹 서버 사이 정보 주고받는 규칙
==> Common Gateway Interface

### CGI 단점
CGI 자체는 어떤 프로그래밍 언어로도 프로그램 개발 가능
- 단점 : 클라이언트 요청에 대해 독립적인 별도 프로세스 생성됨.
  - 요청 많아짐 -> 프로세스도 많아짐 -> 메모리 요구량도 커짐.
    - 그래서 요즘 잘 안씀.

### CGI 대안.
웹 서버에 스크립트 엔진 내장시켜서 별도 프로세스 가동.(Perl, PHP 스크립트)
파이썬 => mod_python -> mod_wsgi
